```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- DATA ----------------
data = {
    "Student": ["Aman", "Riya", "John", "Neha", "Kabir", "Sara"],
    "Marks": [78, 92, 65, 88, 74, 95]
}

df = pd.DataFrame(data)

# ---------------- BAR GRAPH ----------------
plt.figure(figsize=(8,5))
plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks - Bar Graph")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()

# ---------------- PIE CHART ----------------
plt.figure(figsize=(7,7))
plt.pie(df["Marks"], labels=df["Student"], autopct='%1.1f%%')
plt.title("Marks Distribution - Pie Chart")
plt.show()

# ---------------- HISTOGRAM ----------------
plt.figure(figsize=(7,5))
plt.hist(df["Marks"], bins=5, edgecolor='black')
plt.title("Marks Distribution - Histogram")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()

# ---------------- SCATTER PLOT ----------------
plt.figure(figsize=(8,5))
plt.scatter(df["Student"], df["Marks"], color='blue')

plt.title("Student Marks - Scatter Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)

# show values on points
for i, value in enumerate(df["Marks"]):
    plt.text(i, value + 1, str(value), ha='center')

plt.show()
```
